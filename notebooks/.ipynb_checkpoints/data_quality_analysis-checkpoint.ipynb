# Cruise Reservation Data Quality Analysis

This notebook analyzes the quality of a synthetic cruise reservation dataset.

## Objectives
- Identify data quality issues
- Validate business rules
- Analyze booking and payment patterns
- Compute overall data quality score

### Section 1: Setup

```python
import pandas as pd
import matplotlib.pyplot as plt

bookings = pd.read_csv("../data/bookings.csv")
passengers = pd.read_csv("../data/passengers.csv")
payments = pd.read_csv("../data/payments.csv")

print("Datasets loaded.")
```

### Section 2: Basic Overview

```python
print(bookings.head())
print("""--------------------------------------------------""") 
print(passengers.head())
print("""--------------------------------------------------""") 
print(payments.head())
print("""--------------------------------------------------""") 

print("Bookings shape:", bookings.shape)
print("Passengers shape:", passengers.shape)
print("Payments shape:", payments.shape)
```

### Section 3: Missing Data Analysis

```python
missing = bookings.isnull().sum()

missing.plot(kind="bar", title="Missing Values per Column")
plt.show()
```

### Section 4: Duplicate Analysis

```python
duplicates = bookings.duplicated().sum()
print("Duplicate bookings:", duplicates)
```

### Section 5: Business Rule Validation

```python
#Invalid sailing dates
invalid_dates = (bookings["sailing_date"] < bookings["booking_date"]).sum()
print("Invalid sailing dates:", invalid_dates)
```

```python
#Negative fare prices
negative_prices = (bookings["total_fare"] < 0).sum()
print("Negative fares:", negative_prices)
```

### Section 6: Cross-Table Validation

```python
#Orphan passengers 
invalid_passengers = passengers[
    ~passengers["booking_id"].isin(bookings["booking_id"])
]

print("Orphan passengers:", len(invalid_passengers))
```

```python
#Payment mismatch
payment_sum = payments.groupby("booking_id")["payment_amount"].sum()

merged = bookings.set_index("booking_id").join(payment_sum)

overpaid = (merged["payment_amount"] > merged["total_fare"]).sum()

print("Overpaid bookings:", overpaid)
```

### Section 7: Data Quality Score

```python
total_issues = (
    duplicates +
    invalid_dates +
    negative_prices +
    len(invalid_passengers) +
    overpaid
)

total_records = len(bookings)

score = 100 - ((total_issues / total_records) * 100)

print("Data Quality Score:", round(score, 2), "%")
```

### Section 8: Visualization

```python
#Booking distribution
bookings["cabin_type"].value_counts().plot(kind="bar", title="Cabin Type Distribution")
plt.show()
```

```python
#Payment Status
payments["payment_status"].value_counts().plot(kind="bar", title="Payment Status")
plt.show()
```

## Key Findings

- **Missing critical booking data:** A portion of records have null or missing ship values, reducing data completeness and impacting downstream reporting accuracy.
- **Invalid booking timelines:** Instances where sailing_date precedes booking_date indicate data integrity issues that could lead to incorrect scheduling and operational disruptions.
- **Revenue-impacting payment anomalies:** Detection of negative fares and overpayments highlights inconsistencies that can distort revenue reporting and increase the risk of customer disputes.
- **High failed payment rate:** A significant volume of failed transactions suggests potential issues in the payment processing workflow, negatively affecting conversion rates and revenue realization.
- **Low overall data quality:** The calculated Data Quality Score (below 40%) indicates systemic data issues requiring immediate remediation to ensure reliable analytics and decision-making.

## Recommendation

- Implement automated data validation checks at the point of data entry and prior to ingestion into reporting systems.
- Enforce business rules such as valid booking timelines, non-negative pricing, and payment reconciliation constraints.
- Introduce monitoring and alerting for failed payment transactions to improve resolution time and recovery rates.
- Establish data quality metrics and dashboards to continuously track completeness, accuracy, and consistency across booking data.
